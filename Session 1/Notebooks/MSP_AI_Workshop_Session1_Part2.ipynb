```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in data.columns:
  fig, ax = plt.subplots(figsize=(10, 15))
  plt.title(f"{col} Histogram")
  ax.tick_params(axis = 'x', rotation= 90)
  sns.histplot(data = data, x = col, ax = ax)
  plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 15))
plt.title(f"Price Histogram")
ax.tick_params(axis = 'x', rotation= 90)
sns.histplot(data = data, x = 'price', ax = ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
plt.title(f"Price Boxplot")
sns.boxplot(data = data, x = 'price', ax = ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
plt.title("Price vs View")
sns.lineplot(data = data, x = 'price', y = 'view')
plt.show()
```

```python
numeric_col = data.select_dtypes(include='number')
numeric_col.corr()['view']
```

```python
fig , ax = plt.subplots(figsize=(15, 5))
ax = sns.barplot(data = data, x = 'view', y = 'price')
plt.title("Price vs View")
plt.show()
```
