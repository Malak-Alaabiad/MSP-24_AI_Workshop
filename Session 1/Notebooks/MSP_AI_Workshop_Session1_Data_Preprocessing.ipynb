# Data cleaning
1. Splitting (Features and labels)
2. Deal with missing values
3. Scaling or Normlization
4. Drop dublicates
5. Convert text to numbers
6. Split (train test split)

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data = data.reset_index()
```

```python
data = data.drop_duplicates()
```

```python
data = data.drop(['index'], axis = 1)
```

```python
data
```

### Splitting (Features and labels)

```python
x = data.iloc[:, 0:3]
# refers to feactures
```

```python
y = data.iloc[:, 3:4]
# y refers to labels
```

```python
x = pd.get_dummies(x)
```
